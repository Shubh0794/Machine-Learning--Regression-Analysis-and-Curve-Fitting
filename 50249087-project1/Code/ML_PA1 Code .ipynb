```python
print("================================================ TASK 0 ========================================================") 
print('\n')
print('\n')

# TASK 0:

import glmnet_py 
import pandas as pd 
from sklearn import model_selection 
from sklearn.linear_model import RidgeCV 
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LassoCV, Lasso
import sklearn.cross_validation as cv
import numpy as np 
from sklearn.model_selection import train_test_split

datafile =pd.read_csv("ML_SlumpData.csv")

#print(datafile)

X = datafile.iloc[:,[1,2,3,4,5,6,7]]
#print(X)

Y= datafile.iloc[:,9]
#print(Y)

mse_err = list()
alpha_val = list()
coef_val = list()

for i in range (1,10):
    
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.17)
    model= LassoCV(cv=5).fit(X_train,y_train) 
    lasso_model = Lasso(alpha=model.alpha_).fit(X_train,y_train)
    pred = lasso_model.predict(X_test)
    mse_err.append(mean_squared_error(y_test,pred))
    alpha_val.append(model.alpha_)
    coef_val.append(model.coef_)
    
print("MEAN SQUARED ERROR: ",mse_err)        
print('\n')
print("**************")
print('\n')

print("ALPHA VALUES : ",alpha_val)
print('\n')
print("**************")
print('\n')

print("COEFFICIENTS : ",coef_val)
print('\n')
print("**************")
print('\n')

Mean_MSE = np.mean(mse_err)
print ("Average of MSE: ",Mean_MSE)
print('\n')
print("**************")
print('\n')

Std_MSE  = np.std(mse_err)
print ("Standard Deviation : ",Std_MSE)
print('\n')
print("**************")
print('\n') 

#############################################################################################################################    

   
print("================================================ TASK 1.1 ========================================================") 
print('\n') 
print('\n') 
  

#TASK 1.1: 

from matplotlib import pyplot as plt
from sklearn import linear_model
from sklearn.metrics import r2_score


mse_err2 = list()
R2_val = list()
y_pred_ls = list()
y_test_ls = list()



for i in range (1,10):
    
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.17)
    
    lr_model = linear_model.LinearRegression()
    lr_model.fit(X_train,y_train)
    lr_pred= lr_model.predict(X_test)
    R2_val.append(r2_score(y_test,lr_pred))
    mse_err2.append(mean_squared_error(y_test,lr_pred))
    
    y_pred_ls.append(lr_pred) 
    y_test_ls.append(y_test)

    
print("R2 Values : ",R2_val)
print('\n') 
print("**************")
print('\n') 

print("MSE2 Values : ",mse_err2)
print('\n') 
print("**************")
print('\n') 

print(" Y_Predict Values : ",y_pred_ls)
print('\n') 
print("**************")
print('\n') 

print("Y_Test Values : ",y_test_ls)
print('\n') 
print("**************")
print('\n') 

# Getting the index of the Maximum R2 Score and the corresponding y_test and y_predict values

print (" Displaying the index of the Maximum R2 Score and the corresponding y_test and y_predict values")
index_MAX =  R2_val.index(max(R2_val))    
print('\n') 

print("MAXIMUM R2 Score is", R2_val[index_MAX])    
print('\n') 

MAX_y_pred = y_pred_ls[index_MAX]
print("Corresponding Y_Predict is : ", y_pred_ls[index_MAX])
print('\n') 

MAX_y_test = y_test_ls[index_MAX]
print("Corresponding Y_Test is : ", y_test_ls[index_MAX])
print('\n')
print('\n')
print('\n') 
   
plt.scatter(MAX_y_test,MAX_y_pred)

plt.title('Graph : Scatter plot analysis for each Regression ')
plt.xlabel('Maximum Y_Test Values')
plt.ylabel('Maximum Y_Predict Values')
plt.show()

 
#############################################################################################################################    
print('\n')
print('\n')   
print("================================================ TASK 1.2 ========================================================") 
print('\n')
print('\n')  

# TASK 1.2 :

import glmnet_py 
import pandas as pd 
from sklearn import model_selection 
from sklearn.linear_model import RidgeCV, Ridge
from sklearn.metrics import mean_squared_error
#from sklearn.linear_model import LassoCV, Lasso
import sklearn.cross_validation as cv
import numpy as np 
from sklearn.model_selection import train_test_split

datafile =pd.read_csv("ML_SlumpData.csv")

#print(datafile)

X = datafile.iloc[:,[1,2,3,4,5,6,7]]
#print(X)

Y= datafile.iloc[:,9]
#print(Y)

mse_err = list()
alpha_val = list()
coef_val = list()

for i in range (1,10):
    
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.17)
    model= RidgeCV(cv=5).fit(X_train,y_train) 
    ridge_model = Ridge(alpha=model.alpha_).fit(X_train,y_train)
    pred = ridge_model.predict(X_test)
    mse_err.append(mean_squared_error(y_test,pred))
    alpha_val.append(model.alpha_)
    coef_val.append(model.coef_)
    
print("MEAN SQUARED ERROR: ",mse_err)        
print('\n') 
print("**************")
print('\n') 

print("ALPHA VALUES : ",alpha_val)       
print('\n') 
print("**************")
print('\n') 

print("COEFFICIENTS : ",coef_val)
print('\n') 
print("**************")
print('\n') 

Mean_MSE = np.mean(mse_err)
print ("Average of MSE :",Mean_MSE)
print('\n') 
print("**************")
print('\n') 

Std_MSE  = np.std(mse_err)
print ("Standard Deviation : ",Std_MSE)
print('\n') 
print("**************")
print('\n')  


#############################################################################################################################    
 
print("================================================ TASK 1.3 ========================================================") 
print('\n') 
print('\n')   
# TASK 1.3 :


import glmnet_py 
import pandas as pd 
from sklearn import model_selection 
#from sklearn.linear_model import RidgeCV 
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LassoCV, Lasso
import sklearn.cross_validation as cv
import numpy as np 
from sklearn.model_selection import train_test_split

datafile =pd.read_csv("ML_SlumpData.csv")

#print(datafile)

X = datafile.iloc[:,[1,2,3,4,5,6,7]]
#print(X)

Y= datafile.iloc[:,9]
#print(Y)

mse_err = list()
alpha_val = list()
coef_val = list()

for i in range (1,10):
    
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.17)
    model= LassoCV(cv=5).fit(X_train,y_train) 
    lasso_model = Lasso(alpha=model.alpha_).fit(X_train,y_train)
    pred = lasso_model.predict(X_test)
    mse_err.append(mean_squared_error(y_test,pred))
    alpha_val.append(model.alpha_)
    coef_val.append(model.coef_)
    
print("MEAN SQUARED ERROR: ",mse_err)        
print('\n') 
print("**************")
print('\n')  

print("ALPHA VALUES : ",alpha_val)       
print('\n')  
print("**************")
print('\n')  

print("COEFFICIENTS : ",coef_val)
print('\n')  
print("**************")
print('\n')  

Mean_MSE = np.mean(mse_err)
print ("Average of MSE: ",Mean_MSE)
print('\n')  
print("**************")
print('\n')  

Std_MSE  = np.std(mse_err)
print('\n')  
print ("Standard Deviation : ",Std_MSE)
print('\n')  

#############################################################################################################################    
      
print("================================================ TASK 2 ========================================================") 
print('\n')  
  
# TASK 2 :


import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

print(" *******************************************RIDGE*********************************************** ")
print('\n')
print('\n')

Num_alphas = 200
RANGE_of_Alphas = np.logspace(-3, 3, Num_alphas)
Coeff_List = ["Cement","Slag","Fly ash","Water","SP","Coarse Aggr.","Fine Aggr."]

RIDGE_Coeffs = []

for a in RANGE_of_Alphas:
    Regularization_Path = linear_model.Ridge(alpha=a, normalize=True)
    Regularization_Path.fit(X, Y)
    RIDGE_Coeffs.append(Regularization_Path.coef_)

RegPath = plt.gca()

plt.title('Regularization paths : RIDGE ')


RegPath.plot(RANGE_of_Alphas, RIDGE_Coeffs)
RegPath.set_xscale('log')

#Reverse the Axes:

RegPath.set_xlim(RegPath.get_xlim()[::-1]) 

plt.xlabel('Alpha(Penalty) ')
plt.ylabel('Coefficient Weights')
plt.legend(Coeff_List)
plt.show()

# ## ## ## ## ## ## ## ## ## ## ## ## ## ##

print('\n')
print(" *******************************************LASSO*********************************************** ")
print('\n')

Num_alphas = 200
RANGE_of_Alphas = np.logspace(-3, 3, Num_alphas)
Coeff_List = ["Cement","Slag","Fly ash","Water","SP","Coarse Aggr.","Fine Aggr."]

RIDGE_Coeffs = []

for a in RANGE_of_Alphas:
    Regularization_Path = linear_model.Lasso(alpha=a, normalize=True)
    Regularization_Path.fit(X, Y)
    RIDGE_Coeffs.append(Regularization_Path.coef_)

RegPath = plt.gca()

plt.title('Regularization paths : LASSO ')


RegPath.plot(RANGE_of_Alphas, RIDGE_Coeffs)
RegPath.set_xscale('log')

#Reverse the Axes:

RegPath.set_xlim(RegPath.get_xlim()[::-1]) 

plt.xlabel('Alpha(Penalty) ')
plt.ylabel('Coefficient Weights')
plt.legend(Coeff_List)
plt.show()

############################################################# END #########################################################


print("--------------------------------------------------------END---------------------------------------------------------")
```
